```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC

import requests 
import json
from shutil import which
import time
import os
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup
import urllib.request
import itertools

```

```python
chrome_options = Options()


if True:
    chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-infobars")
chrome_options.add_argument("start-maximized")
chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument('--ignore-ssl-errors=yes')
chrome_options.add_argument('--ignore-certificate-errors')

prefs = {"profile.default_content_setting_values.notifications" : 2}
chrome_options.add_experimental_option("prefs",prefs)

chrome_path = r"C:\Users\chaitanya\OneDrive\Documents\Moro\Data scrapping\chromedriver_win32\chromedriver.exe"
global driver
driver = webdriver.Chrome(executable_path=chrome_path, options=chrome_options)
driver.get("http://www.mogg.it/products-list/")


products = WebDriverWait(driver, 5).until(EC.visibility_of_all_elements_located((By.XPATH, "//li[contains(@class ,'w-item')]/figure")))

items = []
for product in products:
    product_details = {}
    
    hero_image = product.find_element(By.XPATH, ".//img")
    product_details['hero_image'] = hero_image.get_attribute('src')

        
    href = product.find_element(By.XPATH, ".//a").get_attribute('href')
    product_details['url'] = href
    
    driver.execute_script("window.open('" +href +"');")
    # switch to new window
    driver.switch_to.window(driver.window_handles[1])
    
    name = driver.find_element(By.XPATH, "//h1[@class = 'entry-title']").text
    product_details['name'] = name
    
    category = driver.find_element(By.XPATH, "//a[@rel = 'tag']").text
    product_details['category'] = category
    
    designer = driver.find_element(By.XPATH, "//h4[@style = 'margin-top: 0px !important']").text
    product_details['designer'] = designer
    
    product_description = driver.find_elements(By.XPATH, "//div[@class = 'portfolio-client-widget widget']/p/em")[0].text
    product_details['product_description'] = product_description
    
    display_dimensions = driver.find_elements(By.XPATH, "//div[@class = 'portfolio-client-widget widget']/p")[-1].text
    product_details['display_dimensions'] = display_dimensions
    
    images = []
    images_path = driver.find_elements(By.XPATH, "//li[contains(@class, 'w-item')]/div/a")
    
    for img in images_path:
        images.append(img.get_attribute('href'))
        
    product_details['images'] = '\n'.join(images)
    
    options = driver.find_elements(By.XPATH, "//div[@class = 'w-text-block']")
    if len(options)>1:
        
        for i in range(0, len(options), 2):
            opt_values = []
            opt_name = options[i].find_element(By.XPATH, ".//p/strong/em").text
            values_path = options[i+1].find_elements(By.XPATH, ".//div/div[2]/a/figure/div/img")
            for val in values_path:
                val_name = val.get_attribute('data-title')
                val_img = val.get_attribute('data-gg-real-image-href')
                opt_values.append(val_name+ ': ' + val_img)
            
            product_details[f'option({opt_name})'] = opt_values
    
    print(product_details)
    items.append(product_details.copy())
    
    driver.close()

    #back to main window
    driver.switch_to.window(driver.window_handles[0])

    
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(items, )
df
```

```python
opt_columns = [i for i in list(df.columns) if 'option' in i]
opt_columns
```

```python
from ast import literal_eval
for i in range(len(df)):
    for col in opt_columns:
        try:
            df[col][i] = literal_eval(df[col][i]) #convert to list type
        except:
            pass
    
        
```

```python
for col in opt_columns:
    df = df.explode(col, ignore_index = True)
```

```python
for i in range(len(df)):
    for col in opt_columns:
        try:
            df[col][i] = df[col][i].replace('<em>', '') #convert to list type
            df[col][i] = df[col][i].replace('</em>', '')
        except:
            pass
    
        
```
