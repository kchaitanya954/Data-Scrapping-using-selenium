```python
def extract(page_source, start_sep, end_sep):
    source=page_source

    result=[]
    tmp=source.split(start_sep)
    for par in tmp:
        if end_sep in par:
            result.append(par.split(end_sep)[0])
    return result

def extract2(string, start, end):
    result = re.search('%s(.*)%s' % (start, end), string).group(1)
    return result
```

```python
category_urls = {'dining tables':'https://www.skovby.com/en-gb/products/dining-tables', \
                'dining chairs':'https://www.skovby.com/en-gb/products/dining-chairs', \
                'storage furniture':'https://www.skovby.com/en-gb/products/storage-furniture', \
                'tv-cabinets':'https://www.skovby.com/en-gb/products/tv-cabinets', \
                'other products':'https://www.skovby.com/en-gb/products/other-products', \
                'coffee tables':'https://www.skovby.com/en-gb/products/coffee-tables', \
                'modo by skovby':'https://www.skovby.com/en-gb/products/modo-by-skovby', \
                'norra by skovby':'https://www.skovby.com/en-gb/products/norra-by-skovby'}
```

```python
chrome_options = Options()
headless = False

if headless:
    chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-infobars")
chrome_options.add_argument("start-maximized")
chrome_options.add_argument("--disable-extensions")


prefs = {"profile.default_content_setting_values.notifications" : 2}
chrome_options.add_experimental_option("prefs",prefs)

# PLease replace your chrome driver path
#         chrome_path = which("chromedriver_win32/chromedriver")
global driver
driver = webdriver.Chrome(executable_path=r"C:\Users\kchai\OneDrive\Documents\Moro\Data scrapping\chromedriver_win32\chromedriver.exe", options=chrome_options)

items = []

empty_item()
for key, value in category_urls.items():
    
    driver.get(value)
    
    try:
        cookies_path = driver.find_elements(By.XPATH, "//button[@class = 'coi-banner__accept']")
        cookies_path[0].click()
    except:
        pass
    
    product_url = driver.find_elements(By.XPATH, "//div[@class = 'productlist-item']/a")
    for product in product_url:
        
        empty_item()
        
        product_details['categories'] = key
        product_details['name'] = product.get_attribute("title")
        
        hero_image = product.find_element(By.XPATH, ".//span/span/span/img[1]")
        product_details['hero_image'] = hero_image.get_attribute("src")
        
        href = product.get_attribute("href")
        driver.execute_script("window.open('" +href +"');")
        print(href)
        product_details['url'] = href
    #     switch to new window
        driver.switch_to.window(driver.window_handles[1])
        time.sleep(1)
        
        product_desc = driver.find_elements(By.XPATH, "//div[@class = 'productpage-text']//p")
        product_description = []
        for desc in product_desc:
            product_description.append(desc.text)
        
        product_details['product_description'] = '\n'.join(product_description)
        
        
        images_path1 = driver.find_elements(By.XPATH, "//div[@class = 'inner']/img")
        images = []
        for img in images_path1:
            images.append(img.get_attribute('src'))
        
        images_path2 = driver.find_elements(By.XPATH, "//div[@class = 'item-image']/img")
        for img in images_path2:
            images.append(img.get_attribute('src'))
            
        product_details['images'] = '\n'.join(images)
        
        videos_path = driver.find_elements(By.XPATH, "//div[@class = 'embed-container']/iframe")
        videos = []
        for vid in videos_path:
            videos.append(vid.get_attribute('src'))
            
        product_details['videos'] = '\n'.join(videos) 
        
        option_name = driver.find_element(By.XPATH, "//div[@class = 'col-xs-12 col-md-4 productpage-materials']/h2")
        option_name = option_name.text
        
        options_path = driver.find_elements(By.XPATH, "//div[@class = 'productpage-materials-item']")
        options = []
        for opt in options_path:
            opt_img = opt.find_element(By.XPATH, ".//div/img").get_attribute('src')
            opt_val = opt.find_element(By.XPATH, ".//img").get_attribute('alt')
            option = opt_val + ": " + opt_img
            options.append(option)
        if len(options) > 0:
            product_details[f'option({option_name})'] = options
        
        
        files_path = driver.find_elements(By.XPATH, "//div[@class = 'productpage-functionality']/div")
        files = []
        for file in files_path:
            file_name = file.find_element(By.XPATH, ".//div[1]")
            try:
                file_name.click()
            except:
                pass
            
            file_name = file_name.text
            
            try:
                file_url = file.find_element(By.XPATH, ".//div[2]//a").get_attribute('href')
#                 if len(file_url) >10:
                temp = file_name + ": " + file_url
                files.append(temp)
            except:
                pass
        files = [data for data in files if '.pdf' in data]
        if len(files) >0:  
            product_details['files'] = '\n'.join(files) 
        else:
            product_details['files'] =  None
        
#         print(product_details)
        items.append(product_details)
        driver.close()

        #back to main window
        driver.switch_to.window(driver.window_handles[0])
```
